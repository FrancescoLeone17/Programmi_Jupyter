# Istogramma della funzione d'onda di prova del VMC

```python
#Plotting script
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

#Histo
nconf = nblk*nstep / 10
    
file_data = np.loadtxt('output.conf.0', usecols=(0))
conf = file_data[:] 


plt.title('')

#plt.ylabel('')
plt.plot(x,(psi[0])**2)
plt.hist(conf,bins=100,density=True)
plt.legend(['Analytical','Trial'], loc='best')
plt.xlabel('x')
#plt.xlim([0, 5000])
#plt.ylim([-7.5825, -7.57])
plt.grid(True)
#plt.savefig('Pictures/histo', dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

#plt.ylabel('')
plt.plot(x,(psi[0])**2)

t = 400

x = np.zeros(t)
pb = np.zeros(t)    
err_pb = np.zeros(t)
file_data = np.loadtxt('../PIMC_PIGS/probability.dat', usecols=(0,1,2))
x = file_data[:,0]
pb = file_data[:,1] 
err_pb = file_data[:,2] 

plt.errorbar(x,pb,yerr=err_pb)
plt.title('PIGS : Funzione di prova uguale a 1')
plt.ylabel('pb')
plt.xlabel('x')
#plt.xlim([0, 5000])
#plt.ylim([-7.5825, -7.57])
plt.legend(['Analytical gs','Sampled'], loc='best')
plt.grid(True)
#plt.savefig('Pictures/Sampling_fzprova_cost', dpi=300)
plt.show()
```
