```python
import numpy as np

def func(x,y):  # Function for statistical uncertainty estimation
    return np.sin(x**2+y**2)

n_train = 10000
n_valid = 1000

# generate synthetic training dataset
x_train = np.zeros((n_train, 2))
y_train = np.zeros((n_train, 1))
# generate synthetic validation data
x_valid = np.zeros((n_valid, 2))
y_valid = np.zeros((n_valid, 1))


sigma = 0
# generate training inputs
np.random.seed(0)
x_train[:,0] = np.random.uniform(-1.5, 1.5, n_train)
x_train[:,1] = np.random.uniform(-1.5, 1.5, n_train)
y_train = np.random.normal(func(x_train[:,0],x_train[:,1]), sigma)

x_valid[:,0] = np.random.uniform(-1.5, 1.5, n_valid)
x_valid[:,1] = np.random.uniform(-1.5, 1.5, n_valid)
y_valid = np.random.normal(func(x_valid[:,0],x_valid[:,1]), sigma)


#z_target = func(x_valid,y_valid) # ideal target function
```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(30, input_shape=(2,),  activation='relu'))
model.add(Dense(25, activation='relu'))
model.add(Dense(30,   activation='relu'))
model.add(Dense(30, activation='elu'))
model.add(Dense(25,  activation='elu'))
model.add(Dense(35,  activation='elu'))
model.add(Dense(1, activation='selu' ))

#model.add(Dense(20, input_shape=(1,),  activation='relu'))
#model.add(Dense(20, activation='relu'))
#model.add(Dense(1, activation='softplus' ))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
#stocastic gradient descent, mean square error
```

```python
# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=200,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
