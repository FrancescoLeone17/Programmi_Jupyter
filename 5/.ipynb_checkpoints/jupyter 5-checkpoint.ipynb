Il programma legge gli input dal file **"input.dat"**, in questo file vengono definiti numero di blocchi, step per ogni blocco, parametro che regola l'accettazione e due parametri importanti:
- **key1** se uguale a 0 viene scelta una transition probability uniforme (nel Metropolis), altrimenti una normale (Gaussian con media in 0 e dev std uguale a 1).
- **key2** se uguale a 0 viene calcolato $<r>_{\psi_{1,0,0}}$, altrimenti $<r>_{\psi_{2,1,0}}$.
Inoltre, la configurazione iniziale che inizializza il Metropolis viene letta dal file **"conf.0"**.

La mossa Metropolis viene svolta dalla funzione **Metropolis**.

Il file di output è **"ave.out"**.

```python
#Plotting script
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

nblk = 200
nstep = 10000

ave = np.zeros(nblk)    
err = np.zeros(nblk)
file_data = np.loadtxt('ex_unif.out', usecols=(0,1,2,3))
ave = file_data[:,2]
err = file_data[:,3]

ave2 = np.zeros(nblk)    
err2 = np.zeros(nblk)
file_data = np.loadtxt('ex_gauss.out', usecols=(0,1,2,3))
ave2 = file_data[:,2]
err2 = file_data[:,3]

x = np.arange(nblk)

plt.errorbar(x,ave,yerr=err)
plt.errorbar(x,ave2,yerr=err2)
plt.title(' Ex Lunghezza di blocco = 10000  ')
plt.xlabel('blocco')
plt.ylabel(' <r>')
plt.legend(['Uniform','Gaussian'], loc='best')
plt.xlim([0, 100])
#plt.ylim([1.48,1.60])
plt.grid(True)
#plt.savefig('Pictures/confronto_ex.png', dpi=300)
plt.show()
```
